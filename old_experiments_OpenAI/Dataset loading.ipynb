```python
from pathlib import Path
from huggingface_hub import snapshot_download
import subprocess

ROOT = Path(r"D:\Users\TimeBound")
ROOT.mkdir(exist_ok=True)

HF_DATASETS = {
    "TempReason": "tonytan48/TempReason",
    "complex-tr": "tonytan48/complex-tr",
    "TCP": "Beanbagdzf/TCP",
    "multiwoz_v22": "pfb30/multi_woz_v22",
}

for name, repo_id in HF_DATASETS.items():
    out_dir = ROOT / name
    print(f"\nDownloading {name} from {repo_id}...")
    snapshot_download(
        repo_id=repo_id,
        repo_type="dataset",
        local_dir=out_dir,
        local_dir_use_symlinks=False,
    )
    print(f"Saved to: {out_dir}")

locomo_dir = ROOT / "locomo"
if not locomo_dir.exists():
    print("\nCloning LoCoMo...")
    subprocess.run(
        ["git", "clone", "https://github.com/snap-research/locomo.git", str(locomo_dir)],
        check=True,
    )
else:
    print("\nLoCoMo already exists.")

print("\nDone.")
```

```python
from pathlib import Path
import zipfile
import requests

ROOT = Path(r"D:\Users\TimeBound")
ROOT.mkdir(exist_ok=True)

locomo_dir = ROOT / "locomo"

def download_and_extract_github_zip(url: str, out_dir: Path):
    zip_path = out_dir.parent / "locomo.zip"

    print(f"Downloading LoCoMo ZIP from: {url}")
    r = requests.get(url, timeout=120)
    r.raise_for_status()

    with zip_path.open("wb") as f:
        f.write(r.content)

    print("Extracting ZIP...")
    with zipfile.ZipFile(zip_path, "r") as z:
        z.extractall(out_dir.parent)

    # GitHub ZIP обычно распаковывается как locomo-main
    extracted = out_dir.parent / "locomo-main"

    if extracted.exists():
        if out_dir.exists():
            print(f"{out_dir} already exists, removing extracted duplicate name not needed.")
        else:
            extracted.rename(out_dir)

    zip_path.unlink(missing_ok=True)
    print(f"LoCoMo saved to: {out_dir}")

if not locomo_dir.exists():
    download_and_extract_github_zip(
        "https://github.com/snap-research/locomo/archive/refs/heads/main.zip",
        locomo_dir,
    )
else:
    print("LoCoMo already exists.")
```
